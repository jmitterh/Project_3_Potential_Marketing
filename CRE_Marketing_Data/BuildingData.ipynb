```python
# extract all files
i = 0
for file in glob.glob(mydir + '/*.zip'):
    i += 1
    zip = ZipFile(file, 'r')
    print(f'Extracting file {i}')
    zip.extractall(mydir)
    zip.close()
    print('Done!')
    print(f"File {i}, extracted: {file}\n")
    time.sleep(1)
#     os.remove(file)
```

```python
for file in glob.glob(mydir + '/*.txt'):
    if "t_business_acct" in file:
        df_business = pd.read_table(file,
                                  sep='\t',
                                  header=None,
                                  names=business_acct, index_col=False, engine= 'python')
        df_business = df_business.append(df_business)
#         os.remove(file)
        print('Added the ' + file + ' into business account txt file')
        print('deleted the file ' + str(file))
    elif "building_other" in file:
        df_other_biz = pd.read_table(file,
                                   sep='\t',
                                     quotechar='"', 
                                     error_bad_lines=False,
#                                   widths=[11, 50, 40, 20, 2, 5, 3, 2, 10, 1, 8, 8, 2, 10, 8, 8, 2, 1, 3, 8, 6],
                                   header=None,
                                   names=other_business, index_col=False, engine= 'python') 
        df_other_biz = df_other_biz.append(df_other_biz)
#         os.remove(file)
        print('Added the ' + file + ' into other business text file')
        print('deleted the file ' + str(file))        
    else:
#         os.remove(file)
        print('File not being used: ' + str(file))
```
